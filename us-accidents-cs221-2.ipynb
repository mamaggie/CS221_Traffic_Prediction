```python
df['Start_Time'] = pd.to_datetime(df['Start_Time'], errors='coerce')
df['End_Time'] = pd.to_datetime(df['End_Time'], errors='coerce')

# Extract year, month, day, hour and weekday
df['Year']=df['Start_Time'].dt.year
df['Month']=df['Start_Time'].dt.strftime('%b')
df['Day']=df['Start_Time'].dt.day
df['Hour']=df['Start_Time'].dt.hour
df['Weekday']=df['Start_Time'].dt.strftime('%a')

# Extract the amount of time in the unit of minutes for each accident, round to the nearest integer
td='Time_Duration(min)'
df[td]=round((df['End_Time']-df['Start_Time'])/np.timedelta64(1,'m'))


df.dtypes
```

```python
neg_outliers=df[td]<=0

# Set outliers to NAN
df[neg_outliers] = np.nan

# Drop rows with negative td
df.dropna(subset=[td],axis=0,inplace=True)
```

```python
period_features = ['Sunrise_Sunset','Civil_Twilight','Nautical_Twilight','Astronomical_Twilight']
df[period_features] = df[period_features].astype('category')
df = pd.get_dummies(df, columns=period_features, drop_first=True)
```

```python
df.head(20)
```

```python
feature_lst=['Severity','Start_Lng','Start_Lat','Distance(mi)','Side','City','County','State','Timezone','Temperature(F)','Humidity(%)',
             'Pressure(in)', 'Visibility(mi)', 'Wind_Direction','Weather_Condition','Amenity','Bump','Crossing','Give_Way','Junction',
             'No_Exit','Railway','Roundabout','Station','Stop','Traffic_Calming','Traffic_Signal','Turning_Loop','Hour',
             'Weekday','Wind_Chill(F)','Wind_Speed(mph)','Precipitation(in)', 
             'Sunrise_Sunset_Night','Civil_Twilight_Night','Nautical_Twilight_Night','Astronomical_Twilight_Night']
df_new = df[feature_lst].copy()
df_new['Severity'] = df_new['Severity'].astype(int)
df_new.info()
```

```python
x_cols = [col for col in df_new.columns if col not in ['Severity'] if df_new[col].dtype=='float64']

labels = []
values = []
for col in x_cols:
    labels.append(col)
    values.append(np.corrcoef(df_new[col].values, df_new.Severity.values)[0,1])
corr_df = pd.DataFrame({'col_labels':labels, 'corr_values':values})
corr_df = corr_df.sort_values(by='corr_values')

ind = np.arange(len(labels))
width = 0.4
fig, ax = plt.subplots(figsize=(12,40))
rects = ax.barh(ind, np.array(corr_df.corr_values.values), color='y')
ax.set_yticks(ind)
ax.set_yticklabels(corr_df.col_labels.values, rotation='horizontal')
ax.set_xlabel("Correlation coefficient")
ax.set_title("Correlation coefficient of the variables")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
y = df_new['Severity'].values
x_cols = [col for col in df_new.columns if col not in ['Severity'] if df_new[col].dtype=='float64']
X= df_new[x_cols]

X_insample, X_test, y_insample, y_test = train_test_split(X, y, test_size = 0.2, random_state= 101)
X_train, X_val, y_train, y_val = train_test_split(X_insample, y_insample, test_size = 0.2, random_state= 101)
```

```python
from sklearn.model_selection import GridSearchCV, KFold, cross_val_predict
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score
from xgboost import XGBClassifier
from xgboost import plot_importance
import xgboost as xgb
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_train = le.fit_transform(y_train)
y_val = le.fit_transform(y_val)
y_test = le.fit_transform(y_test)
```

```python
def fitXgb(sk_model,epochs=300):
    print('Fitting model...')
    sk_model.fit(X_train, 
              y_train, 
              eval_metric="merror", 
              eval_set=[(X_train, y_train), (X_val, y_val)], 
              verbose=20, 
              early_stopping_rounds=20)
    print('Fitting done!')
    
xgb_model = XGBClassifier(max_depth=4, 
                          n_estimators=500, 
                          min_child_weight=1000,  
                          colsample_bytree=0.7, 
                          subsample=0.7, 
                          eta=0.3, 
                          num_class=4,
                          seed=4)

fitXgb(xgb_model)
```

```python
    print('-- Model Report --')
    #print('XGBoost Accuracy: '+str(accuracy_score(xgb_model.predict(X_test), y_test)))
    print('XGBoost F1-Score (Macro): '+str(f1_score(xgb_model.predict(X_test),y_test,average='macro')))
```

```python
from xgboost import plot_importance
def plot_features(booster, figsize):    
    fig, ax = plt.subplots(1,1,figsize=figsize)
    return plot_importance(booster=booster, ax=ax)

plot_features(xgb_model, (10,14))
```

```python
grid = {
 'eta':[0.2, 0.1],
# 'subsample':[0.6, 0.7, 0.8]
}
#metrics to consider: f1_micro, f1_macro, roc_auc_ovr
gsearch1 = GridSearchCV(estimator = xgb_model, param_grid = grid, 
                        scoring='f1_macro',n_jobs=-1,verbose = 10, cv=5)
gsearch1.fit(X_train, y_train)
```

```python
def getTrainScores(gs):
    results = {}
    runs = 0
    for x,y in zip(list(gs.cv_results_['mean_test_score']), gs.cv_results_['params']):
        results[runs] = 'mean:' + str(x) + 'params' + str(y)
        runs += 1
    best = {'best_mean': gs.best_score_, "best_param":gs.best_params_}
    return results, best
```

```python
getTrainScores(gsearch1)
```

```python
#fitting using the best parameters from tuning
xgb_opt = XGBClassifier(max_depth=4, 
                          n_estimators=500, 
                          min_child_weight=1000,  
                          colsample_bytree=0.7, 
                          subsample=0.7, 
                          eta=0.2, 
                          num_class=4,
                          seed=4)

fitXgb(xgb_opt)
```
